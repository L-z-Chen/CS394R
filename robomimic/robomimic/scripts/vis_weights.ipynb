```python
import numpy as np
from matplotlib import pyplot as plt
```

# Uniform sample start

```python
bins = 10
start_weights = [1.0 / bins] * bins
end_weights = [0.02, 0.025, 0.03, 0.04, 0.05, 0.07, 0.10, 0.15, 0.20, 0.35]

start_weights = np.array(start_weights) / sum(start_weights)
end_weights = np.array(end_weights) / sum(end_weights)

alpha = 0.999975

x = np.linspace(0, 2000*100, 100)
y = np.zeros((bins, x.shape[0]))

for i in range(x.shape[0]):
    y[:,i] = (start_weights - end_weights) * (alpha ** x[i]) + end_weights

fig, ax = plt.subplots()
for i in range(bins):
    ax.plot(x, y[i], label='bin_{}'.format(i))

ax.set_xlabel('steps')
ax.set_ylabel('sample probability')
ax.legend()
plt.show()
```

# self-defined initial weights

```python
bins = 10
end_weights = [0.02, 0.025, 0.03, 0.04, 0.05, 0.07, 0.10, 0.15, 0.20, 0.35]
start_weights = end_weights[::-1]

start_weights = np.array(start_weights) / sum(start_weights)
end_weights = np.array(end_weights) / sum(end_weights)

alpha = 0.999975

x = np.linspace(0, 2000*100, 100)
y = np.zeros((bins, x.shape[0]))

for i in range(x.shape[0]):
    y[:,i] = (start_weights - end_weights) * (alpha ** x[i]) + end_weights

fig, ax = plt.subplots()
for i in range(bins):
    ax.plot(x, y[i], label='bin_{}'.format(i))

ax.set_xlabel('steps')
ax.set_ylabel('sample probability')
ax.legend()
plt.show()
```
